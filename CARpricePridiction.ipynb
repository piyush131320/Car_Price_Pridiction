```python
from sklearn.model_selection import train_test_split
```

```python
x = car.drop(['Price'], axis="columns")
y = car['Price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
oye = OneHotEncoder()
oye.fit(x[['name','company','fuel_type']])
```

```python
column_trans= make_column_transformer((OneHotEncoder(categories=oye.categories_),['name','company','fuel_type']), remainder="passthrough")
```

```python
model = LinearRegression()
```

```python
pipeline = make_pipeline(column_trans, model)
```

```python
pipeline.fit(x_train, y_train)
```

```python
y_pred = pipeline.predict(x_test)
```

```python
y_pred
```

```python
r2_score(y_test, y_pred)
```

```python
scores = []
for i in range(1000):
    x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=i)
    model = LinearRegression()
    pipeline = make_pipeline(column_trans, model)
    pipeline.fit(x_train, y_train)
    y_pred = pipeline.predict(x_test)
    scores.append(r2_score(y_test, y_pred))
```

```python
import numpy as np
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=np.argmax(scores))
model=LinearRegression()
pipeline = make_pipeline(column_trans, model)
pipeline.fit(x_train, y_train)
y_pred = pipeline.predict(x_test)
r2_score(y_test,y_pred)
```
